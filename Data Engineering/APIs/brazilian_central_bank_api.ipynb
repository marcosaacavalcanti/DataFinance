# Brazilian Central Bank API

### Link

https://www3.bcb.gov.br/sgspub/localizarseries/localizarSeries.do?method=prepararTelaLocalizarSeries

### Import Libraries

```python
import pandas as pd
```

### Query Function

```python
def central_bank_query(code):
    url = f"http://api.bcb.gov.br/dados/serie/bcdata.sgs.{code}/dados?formato=json"
    data = pd.read_json(url)
    data["date"] = pd.to_datetime(data["data"],dayfirst = True)
    data.set_index("date",inplace = True)
    return data
```

### Example

```python
dict_codes = {
    "ipca":433,
    "igpm":189,
    "selic":432,
    "cdi":12,
    "pnad":24369,
    "international_reserves":13621,
    "pnad_continua":24369,
    "soja":1273
}

for key in dict_codes.keys():
    central_bank_query(dict_codes[key]).plot(title = key)
```
