### Import Libraries

```python
!pip install pandas_datareader

import pandas as pd
import datetime as dt
import pandas_datareader.data as web
```

### Query Function

```python
def federal_reseve_query(code,start_date):
    data = web.DataReader(f"{code}","fred",start_date)
    return data
```

### Example

```python
# Define starting date
start_date = dt.datetime(2000,1,1)

# Get codes
dict_codes = {
    "gdp":"gdp",
    "unemployment_rate":"UNRATE",
    "3-month_bond":"DGS3MO",
    "10-year_bond":"DGS10",
    "m2":"m2"
}

# Call function
for key in dict_codes.keys():
    federal_reseve_query(dict_codes[key],start_date).plot()
```
