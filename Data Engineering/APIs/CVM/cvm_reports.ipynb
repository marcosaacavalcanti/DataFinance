## Import Libraries

```python
import pandas as pd
import numpy as np
```

## Daily Reports

#### Function

```python
def get_reports(year,month):
    url = f'http://dados.cvm.gov.br/dados/FI/DOC/INF_DIARIO/DADOS/inf_diario_fi_{year}{month}.zip'
    reports = pd.read_csv(url,sep = ';')
    reports.columns = ["type",
                       "cnpj",
                       "document_date",
                       "portfolio_total_value",
                       "share_value",
                       "net_worth_value",
                       "daily_raising",
                       "daily_redemption",
                       "quantity_shareholders"]
    return reports
```

#### Example

```python
reports = get_reports("2021","01")
reports
```

## Funds Registration

#### Function

```python
def funds_registration():
    url = "https://dados.cvm.gov.br/dados/FI/CAD/DADOS/cad_fi.csv"
    registration = pd.read_csv(url,sep = ';',encoding='ISO-8859-1')
    registration.columns = ["fund_type",
                            "fund_cnpj",
                            "corporate_name",
                            "register_date",
                            "constitution_date",
                            "cvm_code",
                            "cancellation_date",
                            "status",
                            "status_start_date",
                            "activity_start_date",
                            "activity_end_date",
                            "exercise_start_date",
                            "exercise_end_date",
                            "class",
                            "class_start_date",
                            "fund_rentability",
                            "condominium",
                            "fund_shares",
                            "fund_exclusivity",
                            "long_term_taxation",
                            "target_public",
                            "investment_entity",
                            "performance_fee",
                            "performance_fee_information",
                            "admnistration_fee",
                            "admnistration_fee_information",
                            "net_worth_value",
                            "net_worth_date",
                            "director",
                            "admnistrator_cnpj",
                            "administrator",
                            "personal_juridical_manager_entity",
                            "id_manager",
                            "auditor_cnpj",
                            "auditor",
                            "custodiant_cnpj",
                            "custodiant",
                            "parent_company_cnpj",
                            "parent_company"]
    return registration
```

#### Example

```python
registration = funds_registration()
registration
```
