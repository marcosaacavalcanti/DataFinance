## Import Libraries

```python
import pandas as pd
import numpy as np
```

## Daily Reports

#### Function

```python
def get_reports(year,month):
    url = f'http://dados.cvm.gov.br/dados/FI/DOC/INF_DIARIO/DADOS/inf_diario_fi_{year}{month}.zip'
    return pd.read_csv(url,sep = ';')
```

#### Example

```python
get_reports("2021","01")
```

## Funds Registration

#### Function

```python
def funds_registration(nrows=0):
    url = "https://dados.cvm.gov.br/dados/FI/CAD/DADOS/cad_fi.csv"
    if nrows == 0:
        return pd.read_csv(url,sep = ';',encoding=)
    else:
        return pd.read_csv(url,nrows=nrows,sep=';')
```
