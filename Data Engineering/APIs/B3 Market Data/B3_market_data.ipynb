### Import Data

```python
year = 2022
width_columns = [2,8,2,12,3,12,10,3,4,13,13,13,13,13,13,13,5,18,18,13,1,8,7,13,12,2]
market_data = pd.read_fwf(f"C://Users//letic//Downloads//B3_market_data_{year}.txt", widths = width_columns, header = 0)
```

### Data Treatment

#### Rename columns

```python
market_data.columns = [
    "register_type",
    "date",
    "bdi_code",
    "ticker",
    "market_type",
    "company_name",
    "asset_type",
    "term_market_maturity",
    "currency",
    "open",
    "high",
    "low",
    "mean",
    "last_trade_price",
    "best_buying_order",
    "best_selling_order",
    "quantity_trades",
    "quantity_units",
    "total_volume",
    "strike",
    "prices_correction_indicator",
    "maturity" ,
    "price_factor",
    "strike_points",
    "isin_code",
    "asset_distribution_number"]

market_data
```

#### Drop last line

```python
market_data = market_data[market_data["date"] != "COTAHIST"]
```

#### Dates to index

```python
market_data["date"] = pd.to_datetime(market_data["date"],format = '%Y%m%d')
market_data.set_index(["date"],inplace=True)
market_data
```

#### Comma Adjustments (divide by 100)

```python
comma_list = [
    "open",
    "high",
    "low",
    "mean",
    "last_trade_price",
    "best_buying_order",
    "best_selling_order",
    "total_volume",
    "strike",
    "strike_points"
]
for col in comma_list:
    market_data[col]=[i/100. for i in market_data[col]]
market_data
```

### Market Types
